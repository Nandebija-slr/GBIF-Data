```python
#Hosted Feature Layer for Occurrences 
occurrence_layer_url = gis.content.get("9eb704c66684497e93fad556fc512437")
occurrence_featureLayer = occurrence_layer_url.layers[0]

# === Site polygon coordinates ===
sites = {
    "Teepna_seismic":"13.500330106 -30.5509472859999, 13.500403115 -30.5509018549999, 13.5003967810001 -29.0002910069999, 13.0003337200001 -29.0002924029999, 13.000338355 -30.3437530369999, 13.0010024420001 -30.3443782699999, 13.027530922 -30.3695029319999, 13.0778885330001 -30.4167919409999, 13.0880948730001 -30.4261959359999, 13.111721975 -30.4477612169999, 13.1443532590001 -30.476903297, 13.243359261 -30.562891372, 13.354129803 -30.658249746, 13.500330106 -30.5509472859999",
    # "Roggeveld": '20.50 -32.94, 20.50 -32.93, 20.51 -32.92, 20.51 -32.91, 20.52 -32.91, 20.53 -32.89, 20.54 -32.88, 20.55 -32.871, 20.54 -32.86, 20.54 -32.87, 20.53 -32.87, 20.53 -32.87, 20.53 -32.87, 20.52 -32.88, 20.52 -32.89, 20.51 -32.89, 20.51 -32.89, 20.51 -32.88, 20.51 -32.87, 20.50 -32.87, 20.49 -32.87, 20.49 -32.87, 20.49 -32.88, 20.49 -32.88, 20.49 -32.89, 20.49 -32.90, 20.50 -32.91, 20.49 -32.91, 20.49 -32.91, 20.49 -32.91, 20.48 -32.90, 20.47 -32.89, 20.46 -32.87, 20.45 -32.87, 20.45 -32.87, 20.45 -32.88, 20.45 -32.88, 20.45 -32.89, 20.45 -32.89, 20.45 -32.90, 20.46 -32.91, 20.46 -32.91, 20.47 -32.92, 20.48 -32.93, 20.49 -32.94, 20.50 -32.94',
#     "Rietkloof": '20.54 -32.98, 20.53 -32.98, 20.52 -32.98, 20.51 -32.98, 20.50 -32.99, 20.49 -32.98, 20.47 -32.99, 20.46 -32.99, 20.46 -33.00, 20.46 -33.02, 20.47 -33.03, 20.49 -33.03, 20.50 -33.03, 20.51 -33.04, 20.51 -33.05, 20.52 -33.06, 20.53 -33.05, 20.54 -33.05, 20.54 -33.05, 20.55 -33.06, 20.55 -33.06, 20.55 -33.06, 20.56 -33.05, 20.56 -33.04, 20.56 -33.02, 20.55 -33.02, 20.55 -33.01, 20.55 -33.01, 20.54 -33.00, 20.55 -32.99, 20.55 -32.99, 20.54 -32.98',
#     "Brandvlei": '20.4 -32.96, 20.40 -32.97, 20.40 -32.97, 20.40 -32.98, 20.39 -32.98, 20.39 -32.99, 20.40 -32.99, 20.40 -32.99, 20.41 -32.99, 20.43 -32.98, 20.43 -32.98, 20.44 -32.98, 20.44 -32.98, 20.45 -32.98, 20.45 -32.99, 20.45 -32.99, 20.45 -32.99, 20.46 -32.99, 20.47 -32.98, 20.48 -32.98, 20.48 -32.98, 20.48 -32.98, 20.49 -32.98, 20.49 -32.98, 20.49 -32.97, 20.49 -32.97, 20.49 -32.97, 20.50 -32.97, 20.51 -32.97, 20.51 -32.97, 20.52 -32.97, 20.52 -32.97, 20.52 -32.97, 20.53 -32.97, 20.53 -32.97, 20.53 -32.97,  20.53 -32.96, 20.54 -32.96, 20.54 -32.95, 20.53 -32.94, 20.53 -32.94, 20.52 -32.94, 20.50 -32.94, 20.48 -32.94, 20.47 -32.94, 20.45 -32.93, 20.45 -32.93, 20.45 -32.93, 20.43 -32.93, 20.43 -32.93, 20.41 -32.94, 20.40 -32.94, 20.40 -32.96',
#     "Old Anglo Farm":"29.13 -25.83, 29.13 -25.83, 29.13 -25.83, 29.13 -25.83, 29.13 -25.83, 29.13 -25.83, 29.13 -25.82, 29.13 -25.82, 29.13 -25.82, 29.13 -25.82, 29.12 -25.82, 29.12 -25.82, 29.12 -25.82, 29.12 -25.82, 29.12 -25.81, 29.12 -25.81, 29.12 -25.81, 29.12 -25.81, 29.12 -25.81, 29.12 -25.81,  29.12 -25.81, 29.11 -25.81, 29.11 -25.81, 29.11 -25.81, 29.11 -25.81, 29.11 -25.81, 29.10 -25.81, 29.10 -25.81, 29.10 -25.81, 29.10 -25.81,  29.10 -25.81, 29.10 -25.81, 29.09 -25.81, 29.09 -25.81, 29.09 -25.81, 29.09 -25.80, 29.09 -25.80, 29.09 -25.80, 29.09 -25.80, 29.09 -25.80,  29.09 -25.80, 29.09 -25.80, 29.09 -25.80, 29.09 -25.79, 29.09 -25.79, 29.09 -25.79, 29.09 -25.79, 29.09 -25.79, 29.09 -25.78, 29.09 -25.78,  29.09 -25.78, 29.09 -25.77, 29.09 -25.77, 29.09 -25.77, 29.09 -25.77, 29.08 -25.77, 29.08 -25.77, 29.08 -25.77, 29.08 -25.77, 29.08 -25.77,  29.07 -25.77, 29.07 -25.77, 29.07 -25.77, 29.07 -25.77, 29.06 -25.77, 29.06 -25.77, 29.06 -25.77, 29.06 -25.77, 29.06 -25.77, 29.04 -25.78,  29.04 -25.78, 29.04 -25.78, 29.04 -25.78, 29.04 -25.78, 29.03 -25.78, 29.03 -25.78, 29.03 -25.78, 29.03 -25.78, 29.03 -25.78, 29.03 -25.78,  29.02 -25.79, 29.02 -25.79, 29.02 -25.79, 29.02 -25.80, 29.02 -25.80, 29.02 -25.80, 29.02 -25.80, 29.02 -25.80, 29.02 -25.80, 29.02 -25.80,  29.02 -25.80, 29.02 -25.80, 29.02 -25.81, 29.02 -25.81, 29.02 -25.81, 29.02 -25.81, 29.02 -25.82, 29.03 -25.82, 29.03 -25.82, 29.03 -25.82,  29.03 -25.83, 29.03 -25.83, 29.04 -25.83, 29.04 -25.83, 29.04 -25.83, 29.05 -25.84, 29.05 -25.84, 29.06 -25.84, 29.06 -25.84, 29.06 -25.84,  29.07 -25.84, 29.07 -25.84, 29.07 -25.84, 29.08 -25.84, 29.08 -25.84, 29.08 -25.84, 29.08 -25.84, 29.08 -25.84, 29.09 -25.85, 29.09 -25.85,  29.09 -25.85, 29.09 -25.85, 29.09 -25.85, 29.10 -25.85, 29.10 -25.85, 29.10 -25.85, 29.10 -25.85, 29.10 -25.85, 29.10 -25.85, 29.11 -25.85,  29.11 -25.85, 29.11 -25.85, 29.11 -25.85, 29.11 -25.85, 29.11 -25.85, 29.11 -25.85, 29.11 -25.85, 29.11 -25.85, 29.11 -25.85, 29.11 -25.85,  29.11 -25.85, 29.12 -25.85, 29.12 -25.85, 29.12 -25.85, 29.12 -25.85, 29.12 -25.85, 29.12 -25.85, 29.12 -25.85, 29.12 -25.85, 29.12 -25.85,  29.13 -25.85, 29.13 -25.85, 29.13 -25.85, 29.13 -25.85, 29.13 -25.85, 29.13 -25.84, 29.13 -25.84, 29.13 -25.84, 29.13 -25.84, 29.13 -25.83,  29.13 -25.83, 29.13 -25.83, 29.13 -25.83, 29.13 -25.83, 29.13 -25.83, 29.13 -25.83, 29.13 -25.83, 29.13 -25.83, 29.13 -25.83, 29.13 -25.83"
    # "Saldanha": "18.00875861645906 -32.99254840758528, 18.00820126896474 -32.99393874218066, 18.00999790460558 -32.99439632366625, 18.01029612784219 -32.9968311059212, 18.00895925910626 -32.99799223526082, 18.01327459251077 -33.00006903591553, 18.01621304160983 -33.00132421793633, 18.02139633490305 -33.00448851113106, 18.01832307604079 -33.00635517277656, 18.03800429130396 -33.01754983713893, 18.04054041828858 -33.02951458663566, 18.04218671500745 -33.02942368749586, 18.0432982537946 -33.03445104052958, 18.07506475284891 -33.03017455542229, 18.07385216293967 -33.00667621576502, 18.08196704114151 -33.00719855958191, 18.07259565168742 -32.99268622817694, 18.07037807406643 -32.98812020653222, 18.04974773812995 -32.97314973678095, 18.04783997547932 -32.98050455491447, 18.04697528456271 -32.98152392198183, 18.041869797958 -32.97816713260936, 18.04117510242069 -32.98123223998221, 18.0384389373576 -32.98795934124491, 18.02951673621069 -32.99120968260746, 18.00595673147815 -32.98457773670124, 18.00463014573022 -32.98757183692918, 18.00595323708364 -32.99167817395447, 18.00875861645906 -32.99254840758528",
}

# === Fields for CSV ===
fields = [
    "siteName", "kingdom", "phylum", "class", "order_", "family", "species", "scientificName",
    "infraspecificEpithet", "decimalLatitude", "decimalLongitude", "eventDate", "verbatimLocality",
    "stateProvince", "speciesKey", "basisOfRecord", "collectionCode", "catalogNumber",
    "identifiedBy", "iucnRedListCategory", "coordinateUncertaintyInMeters"
]

# Buffer distances in meters
BUFFER_DISTANCES = {
    "within_5km": 5000,
    "within_10km": 10000,
    "within_15km": 15000,
    "within_20km": 20000
}

# GBIF API URL
GBIF_URL = "https://api.gbif.org/v1/occurrence/search"

# Field mapping for feature layer
field_name_map = {
    "within_5km": "in_5km",
    "within_10km": "in_10km",
    "within_15km": "in_15km",
    "within_20km": "in_20km",
    "order": "order_"
}

# ================== HELPER FUNCTIONS ==================
def create_site_polygon(coord_str):
    coords = []
    for pair in coord_str.split(","):
        parts = pair.strip().split()
        if len(parts) == 2:
            lon = float(parts[0])
            lat = float(parts[1])
            coords.append([lon, lat])
    if coords and coords[0] != coords[-1]:
        coords.append(coords[0])
    return Polygon({"rings": [coords], "spatialReference": {"wkid": 4326}})

def buffer_polygon(site_polygon, distance_m):
    """Buffer a polygon in meters using Shapely and convert back to ArcGIS Polygon."""
    # Convert ArcGIS Polygon to Shapely
    ring_coords = [[float(c) for c in pt] for pt in site_polygon['rings'][0]]
    shapely_poly = ShapelyPolygon(ring_coords)
    
    # Approximate meters to degrees for buffering
    buffered = shapely_poly.buffer(distance_m / 111000)  # ~1 deg latitude = 111 km
    
    # Convert back to ArcGIS Polygon
    buffered_coords = [ [list(coord) for coord in buffered.exterior.coords] ]
    return Polygon({"rings": buffered_coords, "spatialReference": {"wkid": 4326}})

def get_bbox_from_polygon(polygon):
    """Generate bounding box from ArcGIS Polygon."""
    xs = [pt[0] for pt in polygon['rings'][0]]
    ys = [pt[1] for pt in polygon['rings'][0]]
    return {"xmin": min(xs), "ymin": min(ys), "xmax": max(xs), "ymax": max(ys)}

def fetch_gbif_occurrences(bbox_wkt):
    url = "https://api.gbif.org/v1/occurrence/search"
    all_results = []
    offset = 0
    limit=300
    max_records=10000
    min_lon, min_lat = bbox["xmin"], bbox["ymin"]
    max_lon, max_lat = bbox["xmax"], bbox["ymax"]
    
    bbox_wkt = (
        f"POLYGON(({min_lon} {min_lat}, {min_lon} {max_lat}, {max_lon} {max_lat}, {max_lon} {min_lat}, {min_lon} {min_lat}))"
    )

    while True:
        params = {
            "geometry": bbox_wkt,
            "limit": limit,
            "offset": offset,
            "hasCoordinate": "true",
        }
        response = requests.get(url, params=params).json()

        results = response.get("results", [])
        if not results:
            break

        all_results.extend(results)

        offset += limit

        # stop once we hit max_records (10k default)
        if offset >= max_records:
            break

    return all_results

def classify_occurrence_mutual(pt_geom, site_polygon, buffers):
    """Classify point into mutually exclusive buffer zones with 1/0."""
    tags = {}
    pt = Geometry(pt_geom)

    # Within site
    tags['within_site'] = int(pt.within(site_polygon))

    # 5km buffer, exclude site polygon
    tags['in_5km'] = int(pt.within(buffers['within_5km']) and not pt.within(site_polygon))

    # 10km buffer, exclude 5km
    tags['in_10km'] = int(pt.within(buffers['within_10km']) and not pt.within(buffers['within_5km']))

    # 15km buffer, exclude 10km
    tags['in_15km'] = int(pt.within(buffers['within_15km']) and not pt.within(buffers['within_10km']))

    # 20km buffer, exclude 15km
    tags['in_20km'] = int(pt.within(buffers['within_20km']) and not pt.within(buffers['within_15km']))

    # Needed so you donâ€™t filter everything out
    tags['within_20km'] = int(pt.within(buffers['within_20km']))

    return tags

# ================== MAIN SCRIPT ==================
all_occurrences = []

for site_name, coord_str in sites.items():
    print(f" Processing site: {site_name}")

    # 1. Create base site polygon
    site_polygon = create_site_polygon(coord_str)

    # 2. Generate buffers
    buffers = {}
    for dist_key, dist_m in BUFFER_DISTANCES.items():
        buffers[dist_key] = buffer_polygon(site_polygon, dist_m)

    # 3. Get bounding box from 20km buffer to query GBIF
    bbox = get_bbox_from_polygon(buffers["within_20km"])
    gbif_records = fetch_gbif_occurrences(bbox)
    print(f"Retrieved {len(gbif_records)} GBIF records")

    # 4. Process and classify
    for rec in gbif_records:
        try:
            pt_geom = Point({"x": rec["decimalLongitude"], "y": rec["decimalLatitude"], "spatialReference": {"wkid": 4326}})
        except KeyError:
            continue

        tags = classify_occurrence_mutual(pt_geom, site_polygon, buffers)

        if tags["within_20km"] == 0:
            continue

        record_data = {
            "siteName": site_name,
            "scientificName": rec.get("scientificName"),
            "kingdom": rec.get("kingdom"),
            "phylum": rec.get("phylum"),
            "class": rec.get("class"),
            "order_": rec.get("order"),
            "family": rec.get("family"),
            "species": rec.get("species"),
            "decimalLatitude": rec.get("decimalLatitude"),
            "decimalLongitude": rec.get("decimalLongitude"),
            "eventDate": rec.get("eventDate"),
            "basisOfRecord": rec.get("basisOfRecord"),
            "identifiedBy": rec.get("identifiedBy"),
            "catalogNumber": rec.get("catalogNumber"),
            "collectionCode": rec.get("collectionCode"),
            "speciesKey": rec.get("speciesKey"),
            "infraspecificEpithet": rec.get("infraspecificEpithet"),
            "coordinateUncertaintyInMeters": rec.get("coordinateUncertaintyInMeters"),
            "iucnRedListCategory": rec.get("iucnRedListCategory"),
            "verbatimLocality": rec.get("verbatimLocality"),
            "stateProvince": rec.get("stateProvince"),
            "within_site": tags["within_site"],
            "in_5km": tags["in_5km"],
            "in_10km": tags["in_10km"],
            "in_15km": tags["in_15km"],
            "in_20km": tags["in_20km"]
        }
        all_occurrences.append(record_data)
        

# Remove duplicates
df = pd.DataFrame(all_occurrences)
df.drop_duplicates(subset=["siteName", "scientificName", "decimalLatitude", "decimalLongitude"], inplace=True)

# ================== SAVE DATA TO OCCURRENCES CSV ==================
csv_path = "occurrences_data.csv"
df.to_csv(csv_path, index=False)
print(f"CSVs saved")
```

```python
                                                                # PREPARE TO OCCURRENCES DATA TO OCCURRENCES FEATURE_LAYER

# ========== Prepare to push occurrences data to push featureLayer ========

df = df.rename(columns={
    "within_5km": "in_5km",
    "within_10km": "in_10km",
    "within_15km": "in_15km",
    "within_20km": "in_20km",
    "order":"order_"
})

features_to_add = []
for _, row in df.iterrows():
    attributes = row.to_dict()

    # Build geometry (point from lat/lon)
    geometry = {
        "x": attributes["decimalLongitude"],
        "y": attributes["decimalLatitude"],
        "spatialReference": {"wkid": 4326}
    }

    # Remove None/NaN values for AGOL compatibility
    clean_attrs = {k: (None if pd.isna(v) else v) for k, v in attributes.items()}

    features_to_add.append(Feature(geometry=geometry, attributes=clean_attrs))


# ======== For each siteName: delete old + add new =======
site_names = df["siteName"].unique().tolist()

for site in site_names:
    print(f"Updating site: {site}")

    # 1. Delete existing records for that site
    where_clause = f"siteName = '{site}'"
    occurrence_featureLayer.delete_features(where=where_clause)

    # 2. Insert new records for that site
    site_features = [f for f in features_to_add if f.attributes["siteName"] == site]
    if site_features:
        res = occurrence_featureLayer.edit_features(adds=site_features)
        print(f"Inserted {len(site_features)} occurrence records for {site}")
    else:
        print(f"No occurrence records to insert for {site}")
```
